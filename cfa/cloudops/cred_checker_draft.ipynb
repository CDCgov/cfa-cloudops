```python
from cfa.cloudops import CloudClient
import logging

# Optional: Configure logging to see what's happening
logging.basicConfig(level=logging.INFO)

name_dict = {
    "subscription_id": "Subscription ID",
    "tenant_id": "Tenant ID",
    "batch_account": "Batch Account",
    "batch_location": "Batch Location",
    "resource_group": "Resource Group",
    "blob_storage_account": "Blob Storage Account",
    "blob_storage_endpoint": "Blob Storage End Point",
    "container_registry_account": "Container Registry Account",
    "container_registry_server": "Container Registry Server",
    "user_assigned_identity": "Azure Assigned Identity"
}

id_dict = {
    "subscription_id": "ef340bd6-2809-4635-b18b-7e6583a8803b",
    "tenant_id": "48aa478c-32b6-4e41-babc-cc13a9c9b986",
    "batch_account": "cfaprdba",
    "batch_location": "eastus",
    "resource_group": "EXT-EDAV-CFA-PRD",
    "blob_storage_account": "cfaazurebatchprd",
    "blob_storage_endpoint": "https://cfaazurebatchprd.blob.core.windows.net/",
    "container_registry_account": "cfaprdbatchcr",
    "container_registry_server": "https://cfaprdbatchcr.azurecr.io",
    "user_assigned_identity": "/subscriptions/ef340bd6-2809-4635-b18b-7e6583a8803b/resourcegroups/EXT-EDAV-CFA-Network-PRD/providers/Microsoft.ManagedIdentity/userAssignedIdentities/ext-edav-cfa-batch-account"
}

def test_value(var_name, var_value):
    corr_item = id_dict[var_name]
    var_desc = name_dict[var_name]
    print(f'{var_desc}: {var_value} | ' if var_value else f'Missing {var_desc} | ', \
          f'correct' if var_value == corr_item else f'incorrect, should == {corr_item}')
    #print(f'{var_desc} is correct' if var_value == corr_subscription_id else f'{var_desc} should == {corr_subscription_id}')


def test_cloudclient_setup():
    """
    Test CloudClient initialization and print Azure access configuration.
    
    This function safely prints various configuration values without exposing
    any secret keys or sensitive credentials.
    """
    try:
        # Initialize CloudClient with default settings
        print("=" * 60)
        print("Testing CloudClient Azure Setup")
        print("=" * 60)
        
        cc = CloudClient(keyvault = "cfa-predict")
        print("\n✓ CloudClient initialized successfully")
        
        # Print credential method used
        print(f"\nAuthentication Method: {cc.method}")
        
        # Print credential handler details (safe to print - no secrets)
        print("\n--- Credential Configuration ---")
        subscription_id = getattr(cc.cred, 'azure_subscription_id', None)
        test_value("subscription_id", subscription_id)
        
        tenant_id = getattr(cc.cred, 'azure_tenant_id', None)
        test_value("tenant_id", tenant_id)

        batch_account = getattr(cc.cred, 'azure_batch_account', None)
        test_value("batch_account", batch_account)

        batch_location = getattr(cc.cred, 'azure_batch_location', None)
        test_value("batch_location", batch_location)

        resource_group = getattr(cc.cred, 'azure_resource_group_name', None)
        test_value("resource_group", resource_group)

        blob_storage_account = getattr(cc.cred, 'azure_blob_storage_account', None)
        test_value("blob_storage_account", blob_storage_account)
        
        blob_storage_endpoint = getattr(cc.cred, 'azure_blob_storage_endpoint', None)
        test_value("blob_storage_endpoint", blob_storage_endpoint)

        # Access the container registry object and retrieve the user_name
        container_registry = getattr(cc.cred, 'azure_container_registry', None)
        if container_registry:
            container_registry_account = getattr(container_registry, 'user_name', None)
            test_value("container_registry_account", container_registry_account)
            container_registry_server = getattr(container_registry, 'registry_server', None)
            test_value("container_registry_server", container_registry_server)
        else:
            print("Missing Container Registry")

        user_assigned_identity = getattr(cc.cred, 'azure_user_assigned_identity')
        test_value("user_assigned_identity", user_assigned_identity)
        
        # Print client objects (they are initialized successfully if no errors)
        print("\n--- Azure Client Objects ---")
        print(f"Batch Management Client: {type(cc.batch_mgmt_client).__name__}")
        print(f"Compute Management Client: {type(cc.compute_mgmt_client).__name__}")
        print(f"Batch Service Client: {type(cc.batch_service_client).__name__}")
        print(f"Blob Service Client: {type(cc.blob_service_client).__name__}")
        
        # Test credentials by listing subscriptions
        print("\n--- Testing Credentials ---")
        cc.check_credentials()
        
        print("\n" + "=" * 60)
        print("✓ All tests completed successfully!")
        print("=" * 60)

    #TODO make these relevant    
    except Exception as e:
        print(f"\n✗ Error during CloudClient setup: {e}")
        print("\nTroubleshooting tips:")
        print("1. Try initializing with CloudClient(keyvault = 'cfa-predict')")
        print("2. ?")
        print("3. ?")
        raise


if __name__ == "__main__":
    test_cloudclient_setup()
```
